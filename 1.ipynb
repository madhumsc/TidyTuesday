```python
# Create a basic scatter plot using column indices
ggplot(df, aes(x = df.iloc[:,4 ] , y = df.iloc[:,3 ] )) +
  geom_point() 
```

```python

# Create a basic scatter plot using column indices
ggplot(mtcars, aes(x = mtcars[ 1], y = mtcars[6])) +
  geom_point() +
  labs(title = "Scatter Plot of MPG vs. Weight",
       x = colnames(mtcars)[1],  # Label x-axis with the first column name
       y = colnames(mtcars)[6])  # Label y-axis with the sixth column name
```

```python
ggplot(df, aes(x = df[,1], y = df[,2])) +
  geom_point() +
  labs(x = colnames(df)[1], y = colnames(df)[2]) +
  ggtitle("Scatter Plot using Column Indices")
```

```python
p = ggplot(df, aes(x='Year', y=2)) + geom_bar(stat='identity', position='dodge')
print(p)
```

```python
df_filtered = df[df['Year'] > 1800]

bar_chart = (
    ggplot(df_filtered, aes(x='Year')) +
    geom_bar() +
    labs(title='Bar Chart of Species Counts')
)
print(bar_chart)
```

```python
df_filtered = df[df['Year'] > 1800]
histogram_flipper = (
    ggplot(df_filtered, aes(x='Year')) +
    geom_histogram(binwidth=5, fill='skyblue', color='black', alpha=0.7) +
    labs(title='Histogram of Flipper Length')
)
print(histogram_flipper)
```

```python
df_filtered = df[df['Year'] > 1800]
histogram_species = (
    ggplot(df_filtered, aes(x='Year', fill='Entity')) +
    geom_histogram(binwidth=5, color='black', alpha=0.7) +
    facet_wrap('~Entity', scales='Year') +
    labs(title='Histogram of Flipper Length by Species')
)
print(histogram_species)
```

```python
density_plot_flipper = (
    ggplot(df, aes(x='Year')) +
    geom_density(fill='skyblue', color='black', alpha=0.7) +
    labs(title='Density Plot of Flipper Length')
)
print(density_plot_flipper)
```
