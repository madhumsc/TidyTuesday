```python
import pandas as pd
import plotnine as p9
from palmerpenguins import load_penguins
from plotnine import *
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-04-12/indoor_pollution.csv")
df
```

```python
df = df.rename(columns={"Deaths - Cause: All causes - Risk: Household air pollution from solid fuels - Sex: Both - Age: Age-standardized (Percent)": "new_column_name"})
df
```

```python
import pandas as pd
from plotnine import *

# Assuming you have the aggregated DataFrame named aggregated_df

# Create a ggplot chart
ggplot(df, aes(x='Year', y='new_column_name', fill='Continent')) + \
    geom_bar(stat='identity', position='dodge') + \
    labs(title='Summary Statistic by Continent and Year', x='Year', y='Mean Value') + \
    theme_minimal()
```
